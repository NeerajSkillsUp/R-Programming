```r
actual  <- round(runif(100,min=1,max=5),2)
actual
```

```r
head(actual,15)
tail(actual,10)
```

```r
predicted  <- round(runif(100,min=1,max=5),2)
predicted
```

```r
# difference
D = actual - predicted
cat(D)
```

```r
# sum of difference
SOD = sum(actual - predicted)
cat(SOD)
```

```r
# correlation                          # 0 means no relation
C = cor(actual,predicted)              # (-) : inversely proportional and (+) : directly proportional
cat(C)                                 # between -1 and +1
```

```r
# MAE(mean absolute error)
MAE = mean(abs(actual-predicted))
cat(MAE)
```

```r
# Accuracy
accuracy = abs(actual-predicted) > 0.1
accuracy
accuracy = mean(accuracy)*100
accuracy
```

```r
COR = round(cor(actual,predicted),2)
R_square = round(COR*COR,2)
MAE = round(mean(abs(actual-predicted)),2)
Accuracy =round(mean(abs(actual-predicted) > 0.1),2)*100
write.csv(data.frame(COR,R_square,MAE,Accuracy),file="ResultData.csv",row.names=FALSE)
```
